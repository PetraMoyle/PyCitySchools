# PyCitySchools

```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df
```

# District Summary

```python
# total number of schools
unique_schools = school_data_complete_df["school_name"].unique()
total_schools = len(unique_schools)
total_schools

# total number of students
total_students = school_data_complete_df["student_name"].count()
total_students

# total budget
total_budget = school_data_df["budget"].sum()
total_budget

# average math score
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score

# average reading score
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score

# percentage of students with a passing math score (70 or greater)
passing_math_score = school_data_complete_df.loc[school_data_complete_df["math_score"]>=70]
students_passing_math_score = len(passing_math_score)
per_passing_math_score = students_passing_math_score/total_students*100
per_passing_math_score

# percentage of students with a passing reading score (70 or greater)
passing_reading_score = school_data_complete_df.loc[school_data_complete_df["reading_score"]>=70]
students_passing_reading_score = len(passing_reading_score)
per_passing_reading_score = students_passing_reading_score/total_students*100
per_passing_reading_score

# percentage of students who passed math and reading (% Overall Passing)
passing_both = school_data_complete_df.loc[(school_data_complete_df["reading_score"]>=70) & (school_data_complete_df["math_score"]>=70)]
students_passing_both = len(passing_both)
per_students_passing_both = students_passing_both/total_students*100
per_students_passing_both

#Create a dataframe to hold the above results
summary_table_df = pd.DataFrame({"Total Schools" : [total_schools],
                                "Total Students" : [total_students],
                                "Total Budget" : [total_budget],
                                "Average Math Score" : [average_math_score],
                                "Average Reading Score" : [average_reading_score],
                                "% Passing Math" : [per_passing_math_score],
                                "% Passing Reading" : [per_passing_reading_score],
                                "% Overall Passing" : [per_students_passing_both]})

summary_table_df.head(1)
```

# School Summary

```python
# School Name
district_schools = school_data_complete_df.groupby(["school_name"])
#district_schools.count().head(3)

no_of_schools =  len(district_schools)

# School Type
school_type = district_schools['type']

# Total Students
total_students = district_schools["student_name"].count()

# Total School Budget
total_school_budget = district_schools["budget"].mean()

# Per Student Budget
budget_per_student = total_school_budget/total_students

# Average Math Score
avg_math_score = district_schools["math_score"].mean()

# Average Reading Score
avg_reading_score = district_schools["reading_score"].mean()

# % Passing Math
grouped_passing_math = school_data_complete_df.loc[school_data_complete_df["math_score"]>=70].groupby(["school_name"]).count()["student_name"]
per_passing_math = grouped_passing_math/total_students *100

# % Passing Reading
grouped_passing_reading = school_data_complete_df.loc[school_data_complete_df["reading_score"]>=70].groupby(["school_name"]).count()["student_name"]
per_passing_reading = grouped_passing_reading/total_students *100

# % Overall Passing
overall_passing = school_data_complete_df[(school_data_complete_df["math_score"]>=70) & (school_data_complete_df["reading_score"]>=70)].groupby(["school_name"]).count()["student_name"]
per_overall_passing = overall_passing/total_students *100
per_overall_passing

# Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Type" : school_type,
                                  "Total Students" : total_students,
                                  "Total School Budget" : total_school_budget,
                                  "Per Student Budget" : budget_per_student,
                                  "Average Math Score" : avg_math_score,
                                  "Average Reading Score" : avg_reading_score,
                                  "% Passing Math" : per_passing_math,
                                  "% Passing Reading" : per_passing_reading,
                                  "% Overall Passing" : per_overall_passing})

school_summary_df.head()
                                  
```

# Top Performing Schools (By % Overall Passing)

```python
# Top 5 schools based on overall passing rate 
best_performing_schools = school_summary_df.sort_values(by= "% Overall Passing", ascending = False)
best_performing_schools.head()
```

# Bottom Performing Schools (By % Overall Passing)

```python
# Bottom 5 schools based on overall passing rate 
worst_performing_schools = school_summary_df.sort_values(by= "% Overall Passing", ascending = True)
worst_performing_schools.head()
```

# Math Scores by Grade

# Scores by School Spending

```python
# bins & labels
spending_bins = [0, 584, 629, 644, 675] # 5 bins 
spending_labels = ["0<584", "585-629", "630-644", "645-675"] # 4 labels 

# create a new data frame and desired columns 
scores_by_school_spending_df = school_summary_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

# create a new column and group by school size 
scores_by_school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels=spending_labels, include_lowest=True)
scores_by_school_spending_df = scores_by_school_spending_df.groupby("Spending Ranges (Per Student)")
scores_by_school_spending_df.mean()      

                                                              
```

# Scores by School Size

```python
# bins & labels
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# create a new data frame
school_size_df = school_summary_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

# create a new column and group by school size 
school_size_df["School Size"] = pd.cut(school_summary_df["Total Students"], size_bins, labels=size_labels, include_lowest=True)
school_size_df = school_size_df.groupby("School Size")
school_size_df.mean()

```
